```python
order_merge_orderitems = order.merge(order_items, how = 'left', on = 'order_id')
```

```python
order_merge_orderitems.info()
```

```python
order_merge_payment = order_merge_orderitems.merge(order_payments, how = 'left', on = 'order_id')
```

```python
order_merge_payment.head()
```

```python
order_merge_payment.info()
```

```python
order_merge_reviews = order_merge_payment.merge(order_reviews, how = 'left', on = 'order_id')
```

```python
order_merge_reviews.head()
```

```python
order_merge_customer = order_merge_reviews.merge(customers, how = 'left', on = 'customer_id')
```

```python
order_merge_customer.head()
```

```python
order_merge_products = order_merge_customer.merge(products, how = 'left', on = 'product_id')
```

```python
order_merge_products.head()
```

```python
order_merge_sellers = order_merge_products.merge(sellers, how = 'left', on = 'seller_id')
```

```python
order_merge_sellers.head()
```

```python
order_merge_sellers.info()
```

```python
olist_data = order_merge_sellers.merge(product_category_translation, how = 'left', on = 'product_category_name')
```
